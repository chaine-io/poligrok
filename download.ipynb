# Download Law Documents

```python
import pandas as pd
import requests
import os
import re
import filetype
```

```python
events = pd.read_csv('data/events-23_03_2021.csv')
targets = pd.read_csv('data/Climate Laws targets-23_03_2021.csv', encoding='unicode_escape')
```

Sanitise CSVs manually by removing \N chars

```python
laws = pd.read_csv('data/legislations-23_03_2021.csv', encoding= 'unicode_escape')
full_documents = pd.read_csv('data/laws_and_policies_22032021.csv', encoding='unicode_escape')
```

Extract all URLs per row

```python
full_documents['doc_title'] = full_documents['ï»¿Title'] + "_" + full_documents['Geography']
```

```python
filenames = []
for i, row in full_documents.iterrows():
    try:
        url = re.search("(?P<url>https?://[^\s]+)", row['Documents']).group("url")
        r = requests.get(url)
        with open(f"data/documents/{row['doc_title']}.pdf", 'wb') as f:
            f.write(r.content)
        filenames.append(f"{row['doc_title']}.pdf")
        print(f"Downloaded {row['doc_title']}")
    except:
        filenames.append(None)
        print(f"Failed to download {row['doc_title']}")
full_documents['filename'] = filenames
```

```python
print(f"Number of laws: {len(full_documents)}")
print(f"Number of laws with document links: {len(full_documents[~full_documents['Documents'].isna()])}")
print(f"Of of laws with file links, percentage successfully downloaded: {round(100* len(full_documents[(~full_documents['Documents'].isna()) & (~full_documents['filename'].isna())]) / len(full_documents[~full_documents['Documents'].isna()]), 2)}%")
print(f"Files downloaded: {len(os.listdir('data/documents'))}")
print(f"Overall percentage of laws with files downloaded: {round(100 * len(os.listdir('data/documents')) / len(full_documents), 2)}%")
```
